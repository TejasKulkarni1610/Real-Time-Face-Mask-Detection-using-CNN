```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2, VGG16, ResNet50, InceptionV3
from tensorflow.keras.layers import Dropout, Flatten, Dense, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os 


IMG_DIMS = (224, 224)
DATASET_PATH = r"E:\ANN Project\dataset" 
MODEL_FILENAME_PREFIX = "local_mask_detector_" 


# --- DATA GENERATOR SETUP ---

def create_generators(path, batch_size, target_size):
    """Creates training and validation generators using flow_from_directory."""
    
    print("[INFO] Setting up data generators...")
    
    # Training: Rescale + Augmentation
    train_datagen = ImageDataGenerator(
        rescale=1./255, 
        validation_split=0.20,
        rotation_range=20, 
        zoom_range=0.15, 
        width_shift_range=0.2,
        height_shift_range=0.2, 
        shear_range=0.15, 
        horizontal_flip=True,
        fill_mode="nearest"
    )

    # Validation/Test: Rescale ONLY (No augmentation)
    test_datagen = ImageDataGenerator(
        rescale=1./255, 
        validation_split=0.20
    )

    # 1. Create the Training Generator
    train_generator = train_datagen.flow_from_directory(
        path,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='training', 
        shuffle=True
    )

    # 2. Create the Validation/Testing Generator
    test_generator = test_datagen.flow_from_directory(
        path,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation', 
        shuffle=False
    )
    
    return train_generator, test_generator


# --- TRANSFER LEARNING MODEL CONSTRUCTION ---

def build_transfer_model(base_model_name):
    """Initializes the base model and attaches a classification head."""
    print(f"[INFO] Initializing {base_model_name}...")

    input_shape = (IMG_DIMS[0], IMG_DIMS[1], 3)
    BaseModel = None

    if base_model_name == 'VGG16':
        BaseModel = VGG16
    elif base_model_name == 'ResNet50':
        BaseModel = ResNet50
    elif base_model_name == 'InceptionV3':
        BaseModel = InceptionV3
    elif base_model_name == 'MobileNetV2':
        BaseModel = MobileNetV2
    else:
        raise ValueError("Invalid model name specified.")

    # Load the base model with pre-trained ImageNet weights, excluding the top layer
    baseModel = BaseModel(weights="imagenet", include_top=False,
                            input_tensor=Input(shape=input_shape))

    # Construct the classification head
    headModel = baseModel.output
    headModel = tf.keras.layers.GlobalAveragePooling2D()(headModel) 
    headModel = Dense(128, activation="relu")(headModel)
    headModel = Dropout(0.5)(headModel)
    headModel = Dense(2, activation="softmax")(headModel) # 2 classes (mask/no mask)

    model = Model(inputs=baseModel.input, outputs=headModel)

    # Freeze the base model layers
    for layer in baseModel.layers:
        layer.trainable = False

    return model

# --- PLOTTING FUNCTIONS ---

def plot_confusion_matrix_heatmap(actual_labels, predIdxs, model_name, optim_type, class_names):
    """Plots the confusion matrix as a heatmap."""
    
    cm = confusion_matrix(actual_labels, predIdxs)
    cm_df = pd.DataFrame(cm, index=class_names, columns=class_names)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', 
                linewidths=.5, linecolor='black', cbar=False) 
    
    plt.title(f'Confusion Matrix: {model_name} ({optim_type})')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    plot_filename = f"cm_heatmap_{model_name}_{optim_type}.png"
    plt.tight_layout()
    plt.savefig(plot_filename)
    plt.close() 

def plot_training_history(H, model_name, optim_type):
    """Plots the training loss/accuracy history per epoch."""
    
    N = len(H.history["loss"])
    plt.style.use("ggplot")
    plt.figure(figsize=(12, 6))

    # Plot Loss
    plt.subplot(1, 2, 1)
    plt.plot(np.arange(0, N), H.history["loss"], label="Train Loss")
    plt.plot(np.arange(0, N), H.history["val_loss"], label="Val Loss")
    plt.title(f"Loss: {model_name} ({optim_type})")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(np.arange(0, N), H.history["accuracy"], label="Train Accuracy")
    plt.plot(np.arange(0, N), H.history["val_accuracy"], label="Val Accuracy")
    plt.title(f"Accuracy: {model_name} ({optim_type})")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    
    plot_filename = f"training_plot_{model_name}_{optim_type}.png"
    plt.tight_layout()
    plt.savefig(plot_filename)
    plt.close() 

def plot_comparative_metrics(results_df):
    """Generates a bar plot comparing Accuracy, F1-Score, Precision, Recall, and Training Time."""
    
    models = results_df['Model']
    metrics_df = results_df[['Accuracy (%)']].copy()
    metrics_df[['F1-Score', 'Precision', 'Recall']] = results_df[['F1-Score', 'Precision', 'Recall']] * 100
    
    x = np.arange(len(models))
    width = 0.18 

    fig, ax1 = plt.subplots(figsize=(14, 7))
    
    # Plotting Classification Metrics on the primary axis (ax1)
    ax1.bar(x - 2 * width, metrics_df['Accuracy (%)'], width, label='Accuracy (%)', color='skyblue')
    ax1.bar(x - width, metrics_df['F1-Score'], width, label='F1-Score (%)', color='lightcoral')
    ax1.bar(x, metrics_df['Precision'], width, label='Precision (%)', color='lightgreen')
    ax1.bar(x + width, metrics_df['Recall'], width, label='Recall (%)', color='gold')
    
    ax1.set_ylabel('Performance Metric (%)', color='black', fontweight='bold')
    ax1.set_xlabel('Model Architecture', fontweight='bold')
    ax1.set_title('Comparative Performance Metrics vs. Training Time', fontweight='bold')
    ax1.set_xticks(x)
    ax1.set_xticklabels(models)
    ax1.set_ylim(0, 100) 

    # Create a secondary axis for Training Time (ax2)
    ax2 = ax1.twinx()
    ax2.plot(x, results_df['Training Time (min)'], color='forestgreen', marker='o', linestyle='--', linewidth=3, label='Training Time (min)')
    ax2.set_ylabel('Training Time (min)', color='forestgreen', fontweight='bold')
    ax2.tick_params(axis='y', labelcolor='forestgreen')
    ax2.grid(False) 

    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines + lines2, labels + labels2, loc='upper left', ncol=3)

    plot_filename = "comparative_metrics_plot.png"
    plt.tight_layout()
    plt.savefig(plot_filename)
    plt.close(fig)

def plot_comparative_box_plot(results_df):
    """Generates a box plot comparing key metrics across all models."""
    
    melted_df = results_df[['Model', 'Accuracy (%)', 'F1-Score', 'Precision', 'Recall']].copy()
    melted_df[['F1-Score', 'Precision', 'Recall']] = melted_df[['F1-Score', 'Precision', 'Recall']] * 100
    
    melted_df = melted_df.melt(id_vars='Model', 
                               value_vars=['Accuracy (%)', 'F1-Score', 'Precision', 'Recall'],
                               var_name='Metric', 
                               value_name='Value (%)')
    
    plt.figure(figsize=(12, 7))
    sns.boxplot(x='Metric', y='Value (%)', hue='Model', data=melted_df, palette='viridis')
    
    plt.title('Distribution of Classification Metrics by Model')
    plt.xlabel('Metric')
    plt.ylabel('Value (%)')
    plt.legend(title='Model', loc='lower left')
    plt.grid(True, axis='y', linestyle='--')
    
    plot_filename = "comparative_box_plot.png"
    plt.tight_layout()
    plt.savefig(plot_filename)
    plt.close()


def display_final_plots(results_df):
    """Loads and displays the key saved plots (2x2 grid)."""
    
    if results_df.empty:
        print("[INFO] No results to display plots.")
        return
        
    best_model_row = results_df.sort_values(by='Accuracy (%)', ascending=False).iloc[0]
    model_name = best_model_row['Model']
    optim_type = best_model_row['Optimizer']
    
    comparative_bar_plot_path = "comparative_metrics_plot.png"
    comparative_box_plot_path = "comparative_box_plot.png"
    best_training_plot_path = f"training_plot_{model_name}_{optim_type}.png"
    best_cm_plot_path = f"cm_heatmap_{model_name}_{optim_type}.png"
    
    print("\n[INFO] Displaying FINAL Plots (2x2 Grid)...")
    
    fig = plt.figure(figsize=(20, 15))
    
    # 1. Comparative Bar Plot
    if os.path.exists(comparative_bar_plot_path):
        ax1 = fig.add_subplot(2, 2, 1)
        img1 = plt.imread(comparative_bar_plot_path)
        ax1.imshow(img1)
        ax1.set_title("Comparative Metrics (Bar Chart)")
        ax1.axis('off') 

    # 2. Comparative Box Plot
    if os.path.exists(comparative_box_plot_path):
        ax2 = fig.add_subplot(2, 2, 2)
        img2 = plt.imread(comparative_box_plot_path)
        ax2.imshow(img2)
        ax2.set_title("Comparative Metrics (Box Plot)")
        ax2.axis('off')

    # 3. Best Model Training History Plot
    if os.path.exists(best_training_plot_path):
        ax3 = fig.add_subplot(2, 2, 3)
        img3 = plt.imread(best_training_plot_path)
        ax3.imshow(img3)
        ax3.set_title(f"Training History: {model_name}")
        ax3.axis('off')

    # 4. Best Model Confusion Matrix Heatmap
    if os.path.exists(best_cm_plot_path):
        ax4 = fig.add_subplot(2, 2, 4)
        img4 = plt.imread(best_cm_plot_path)
        ax4.imshow(img4)
        ax4.set_title(f"Confusion Matrix: {model_name}")
        ax4.axis('off')
    
    plt.tight_layout()
    plt.show()


# --- TRAINING AND EVALUATION ---

def run_experiment(model_name, lr, optim_type, bs, train_gen, test_gen):
    tf.keras.backend.clear_session() 
    start_time = time.time()

    # 1. Build Model
    model = build_transfer_model(model_name)

    # 2. Select Optimizer
    if optim_type == 'Adam':
        opt = Adam(learning_rate=lr)
    elif optim_type == 'SGD':
        opt = SGD(learning_rate=lr, momentum=0.9)
    else:
        raise ValueError("Invalid optimizer specified.")

    # 3. Compile
    model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])

    # 4. Early Stopping
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    print(f"[INFO] Training {model_name} (LR: {lr}, Optim: {optim_type}, BS: {bs})...")

    # 5. Training (Epochs set low for quick test)
    H = model.fit(
        train_gen,
        steps_per_epoch=train_gen.samples // bs, 
        validation_data=test_gen,
        validation_steps=test_gen.samples // bs,
        epochs=20,
        callbacks=[early_stop]
    )

    end_time = time.time()
    training_time = (end_time - start_time) / 60.0 

    # 6. Evaluation
    print(f"[INFO] Evaluating {model_name}...")
    
    test_gen.reset() 
    
    steps_for_eval = test_gen.samples // bs
    pred = model.predict(test_gen, steps=steps_for_eval)
    
    # Get actual labels and predictions
    actual_labels = test_gen.classes[:steps_for_eval * bs]
    predIdxs = np.argmax(pred, axis=1)

    # Determine class names for the Confusion Matrix
    class_names = sorted(test_gen.class_indices.keys())

    # Calculate metrics
    accuracy = np.mean(predIdxs == actual_labels) * 100
    f1 = f1_score(actual_labels, predIdxs, average='weighted')
    precision = precision_score(actual_labels, predIdxs, average='weighted')
    recall = recall_score(actual_labels, predIdxs, average='weighted')

    # Save the trained model
    model_path = f"{MODEL_FILENAME_PREFIX}{model_name}_{optim_type}.h5"
    model.save(model_path, save_format="h5") 

    # Plot the Confusion Matrix
    plot_confusion_matrix_heatmap(actual_labels, predIdxs, model_name, optim_type, class_names)
    
    result = {
        'Model': model_name,
        'Optimizer': optim_type,
        'Learning Rate': lr,
        'Batch Size': bs,
        'Accuracy (%)': round(accuracy, 2),
        'F1-Score': round(f1, 4),
        'Precision': round(precision, 4),
        'Recall': round(recall, 4),
        'Training Time (min)': round(training_time, 2)
    }

    print(f"[{model_name} Results] Accuracy: {result['Accuracy (%)']}%, F1: {result['F1-Score']}, Precision: {result['Precision']}, Recall: {result['Recall']}")
    print("-" * 50)

    return result, H


# --- MAIN EXECUTION BLOCK (Comparative Loop) ---

if __name__ == '__main__':

    # Define the experiment configurations
    experiment_configs = [
        {'model': 'VGG16', 'lr': 0.001, 'optim': 'Adam', 'bs': 64},
        {'model': 'ResNet50', 'lr': 0.001, 'optim': 'Adam', 'bs': 64},
        {'model': 'InceptionV3', 'lr': 0.001, 'optim': 'SGD', 'bs': 64},
        {'model': 'MobileNetV2', 'lr': 0.001, 'optim': 'Adam', 'bs': 64},
    ]

    all_results = []
    
    print("\n\n#####################################################")
    print("## STARTING COMPARATIVE TRANSFER LEARNING EXPERIMENTS ##")
    print("#####################################################\n")

    for config in experiment_configs:
        
        # 1. Create generators specific to the required batch size (bs)
        try:
            train_gen, test_gen = create_generators(DATASET_PATH, config['bs'], IMG_DIMS)
        except Exception as e:
            print(f"\n[CRITICAL ERROR] Failed to create data generators. Check if DATASET_PATH ('{DATASET_PATH}') is correct and contains subdirectories for classes.")
            print(f"Error details: {e}\n")
            continue 

        try:
            # 2. Run experiment and get results/history
            result, history = run_experiment(
                config['model'], config['lr'], config['optim'], config['bs'],
                train_gen, test_gen
            )
            all_results.append(result)
            
            # 3. Plot the per-epoch training history (saves to file)
            plot_training_history(history, config['model'], config['optim'])
            
        except Exception as e:
            print(f"[ERROR] Failed to run experiment for {config['model']} with config {config}. Error: {e}")

    # Final Comparison and Plotting
    if all_results:
        results_df = pd.DataFrame(all_results)
        best_model_row = results_df.sort_values(by='Accuracy (%)', ascending=False).iloc[0]

        print("\n\n#####################################################")
        print("## FINAL COMPARATIVE RESULTS ##")
        print("#####################################################")
        print(results_df.to_markdown(index=False))

        # Generate and save the final comparative plots
        plot_comparative_metrics(results_df) # Bar plot
        plot_comparative_box_plot(results_df) # Box Plot

        # Display all saved plots interactively at the end
        display_final_plots(results_df) 

        print("\n\n#####################################################")
        print("## CONCLUSION: BEST PERFORMING MODEL ##")
        print("#####################################################")
        print(f"The best performing model is **{best_model_row['Model']}** with the following configuration:")
        print(f" - Accuracy: **{best_model_row['Accuracy (%)']}%**")
        print(f" - F1-Score: {best_model_row['F1-Score']}")
        print(f" - Precision: {best_model_row['Precision']}")
        print(f" - Recall: {best_model_row['Recall']}")
        print(f" - Training Time: {best_model_row['Training Time (min)']} minutes")
        print(f" - Configuration: Optimizer={best_model_row['Optimizer']}, LR={best_model_row['Learning Rate']}, BS={best_model_row['Batch Size']}")

    print("\n[INFO] All experiments complete. Trained models, plots, and heatmaps saved to your local directory.")
```
